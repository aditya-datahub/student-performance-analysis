```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
```

```python
df.isnull().sum()
```

```python
# Drop column - Unnamed: 0   

df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].replace({'05-Oct': '5-10'})
```

```python
df.head()
```

```python
plt.figure(figsize=(5, 5))
ax = sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# from the above chart we have analysed taht : the number of feamles in the data is more than the number of males 
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
# Heatmap based on ParentEduc and their respective subject mean scores:

plt.figure(figsize= (5,5))
sns.heatmap(gb, annot =True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded that the education of the parents have a good impact on their childrens.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb1) 
```

```python
# Heatmap based on ParentMaritalStatus and their respective subject mean scores:

plt.figure(figsize= (5,5))
sns.heatmap(gb1, annot =True)
plt.title("Relationship between Parent's MaritalStatus and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded that there is no/negligible impact on the student's score due to their parent's marital status
```

# Distribution of Ethnic Groups

```python
# Count each ethnic group
ethnic_counts = df['EthnicGroup'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    ethnic_counts.values,
    labels=ethnic_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel')
)
plt.title('Distribution of Ethnic Groups')
plt.show()
```
